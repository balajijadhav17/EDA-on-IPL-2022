```python
match_counts = df["match_winner"].value_counts()
plt.figure(figsize=(10, 6))
plt.bar(match_counts.index, match_counts.values, color='cyan')
plt.title("Number of Matches Won in by each team in IPL 2022")
plt.xlabel("Teams")
plt.ylabel("No. Matches Won")
plt.show()
```

```python
total_runs_by_player = df.groupby("top_scorer")["highscore"].sum()
top_5_scorers = total_runs_by_player.sort_values(ascending=False).head(5)
plt.figure(figsize=(10, 6))
plt.bar(top_5_scorers.index, top_5_scorers.values, color='blue')
plt.title("Top 5 Players with Most Runs in IPL 2022")
plt.xlabel("Players")
plt.ylabel("Total Runs") 
plt.show()
```

```python
matches_by_venue = df["venue"].value_counts()
plt.figure(figsize=(10, 8))
plt.pie(matches_by_venue.values, labels=matches_by_venue.index, 
        autopct='%1.1f%%', startangle=140, textprops={'fontsize': 12})
plt.title("Number of Matches Played at Each Ground", fontsize=16)
plt.show()
```

```python
won_by_counts = df['won_by'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(won_by_counts, labels=won_by_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Match Wins by Method")
plt.legend(title="Won By", loc="upper left")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.plot(df.index, df["first_ings_wkts"], color="blue", marker="o")
plt.xlabel("Match Index")
plt.ylabel("Wickets")
plt.title("First Innings")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df.index, df["second_ings_wkts"], color="cyan", marker="o")
plt.xlabel("Match Index")
plt.ylabel("Wickets")
plt.title("Second Innings Wickets")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df.index, df["first_ings_wkts"], color="blue", marker="o")
plt.plot(df.index, df["second_ings_wkts"],  color="cyan", marker="o")
plt.xlabel("Match Index")
plt.ylabel("Wickets")
plt.title("First Innings vs Second Innings Wickets")
plt.show()
```

```python
top_players = df['player_of_the_match'].value_counts().head(5)
plt.figure(figsize=(10, 6))
plt.bar(top_players.index, top_players.values, color='purple')
plt.xlabel('Player')
plt.ylabel('Number of Times Awarded')
plt.title('Top 5 Players with Most Player of the Match Awards')
plt.show()
```

```python
top_players = df.groupby("team1")["first_ings_score"].max().head(5)
plt.figure(figsize=(10, 6))
plt.bar(top_players.index, top_players.values, color='blue')
plt.xlabel('Team')
plt.ylabel('Score')
plt.title('Top 5 Team with Most runs in 1st inning')
plt.show()
```
